### Here we will clean up the data

```python
import pandas as pd
from IPython.display import display
```

Let's begin...

```python
astma = pd.read_csv("../data/asthma_disease_data.csv")
pd.set_option('display.max_columns', None)

astma.DoctorInCharge.value_counts()
# We see that all doctors in charge are confidential, so they are useless to us.

astma.drop(columns=["DoctorInCharge", "PatientID"], inplace=True)

astma[(astma.Gender != 1) & (astma.Gender != 0)] # Just in case lol

# For more clarity, I will map the nomenclature columns to their category names.
astma.Gender = astma.Gender.astype('category').map({0: 'Male', 1: 'Female'})
astma.Ethnicity = astma.Ethnicity.astype('category').map({0: 'Caucasian', 1: 'African American', 2: 'Asian', 3: 'Other'})
astma.EducationLevel = astma.EducationLevel.astype('category').map({0: 'None', 1: 'High School', 2: 'Bachelor''s', 3: 'Higher'})

astma.describe(), astma.isnull().sum()
# The rest of the data seems to be good, no separation from the dataset description in kaggle.

# I will save it as a new file so I can acccess it from other notebooks later.
astma.to_csv("../data/asthma_disease_data_cleaned.csv", index=False)
```
