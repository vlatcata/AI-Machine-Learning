## Data exploration of the astma_disease dataset.

First, let's import our necessary libraries, starting with pandas, we will import more if needed.

I will also import display from IPYthon so I can display the result of this notebook in other ones by using %run "notebook"

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display
```

#### Then... let's read the dataset.

```python
astma = pd.read_csv("../data/asthma_disease_data.csv")

astma
```

Now let's explore a little by visualizing some parts of the dataset.

```python
plt.hist(astma.Age, rwidth=0.9)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Age Distribution of Asthma Patients")
plt.show()
```

Actually, instead of going around making 15 visualizations, I will just make a heatmap showing the correlations...

```python
# Need to remove the patientId and DoctorInCharge because they cannot be used for correlations (non-numeric)
astma_corr = astma.drop(columns=["DoctorInCharge", "PatientID"], inplace=False)

plt.figure(figsize=(20, 11))
sns.heatmap(astma_corr.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation heatmap")
plt.show()
```
