```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from IPython.display import display
```

```python
astma = pd.read_csv('../data/asthma_disease_data_cleaned.csv')

# Pairwise plots for key numeric variables and target (Diagnosis)
sns.pairplot(astma, vars=['Age', 'BMI', 'LungFunctionFEV1', 'LungFunctionFVC'], hue='Diagnosis')
plt.show()
```

```python
# Check if there are any significant differences in BMI by Gender
sns.boxplot(x='Gender', y='BMI', data=astma)
plt.title('BMI by Gender')
plt.show()
```

```python
# Cross-tabulations between Education level and Diagnosis
# Very interesting, seems like the higher the education, the more likely the person is to be diagnosed.
pd.crosstab(astma.EducationLevel, astma.Diagnosis, normalize='index')
```
